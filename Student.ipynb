```python
emp_records = [
    (1, 'Steven', (95, 90, 60, 65, 77, 80)),
    (2, 'Nick',   (95, 93, 60, 65, 60, 81)),
    (3, 'Peter',  (77, 90, 34, 76, 77, 80)),
    (4, 'Dan',    (95, 58, 60, 65, 66, 89)),
    (5, 'Tim',    (23, 55, 56, 43, 56, 45)),
    (6, 'Dave',   (0, 35, 55, 76, 77, 80)),
    (7, 'John',   (92, 'Ab', 60, 65, 77, 80)),
    (8, 'Jacob',  (98, 89, 60, 65, 77, 80)),
    (9, 'Chris',  (97, 98, 96, 99, 97, 89)),
]
```

```python
total_list = []
for rec in emp_records:
    total_list.append((rec[0], rec[1], sum([score for score in rec[2] if isinstance(score, int)])))
print(total_list)
```

```python
passed = []

for rec in emp_records:
    if not isinstance(rec[0], int):
        continue

    marks = rec[3] if len(rec) >= 4 else rec[2]

    if all(isinstance(m, int) and m >= 35 for m in marks):
        passed.append((rec[1], sum(m for m in marks if isinstance(m, int))))

max_total = max(t[1] for t in passed)
toppers = [name for name, total in passed if total == max_total]

print("Toppers:", toppers)
```

```python
min_total = 9999
last_grade = []

for _, name, marks in emp_records:
    if all(isinstance(m, int) and m >= 35 for m in marks):
        total = sum(marks)
        if total < min_total:
            min_total, last_grade = total, [name]
        elif total == min_total:
            last_grade.append(name)

print(last_grade)
```

```python
failures = []

for _, name, marks in emp_records:
    if not all(isinstance(m, int) and m >= 35 for m in marks):
        failures.append(name)

print(failures)
```

```python
totals = []

for _, name, marks in emp_records:
    if all(isinstance(m, int) and m >= 35 for m in marks):
        totals.append([name, sum(marks)])

print(totals)
```

```python
passed = [(name, marks) 
          for _, name, marks in emp_records
          if all(isinstance(m, int) and m >= 35 for m in marks)]
```

```python
ai_scores = [(name, marks[1]) for name, marks in passed]

max_ai = max(s for _, s in ai_scores)
min_ai = min(s for _, s in ai_scores)

top_in_ai   = [name for name, s in ai_scores if s == max_ai]
least_in_ai = [name for name, s in ai_scores if s == min_ai]

print(top_in_ai, least_in_ai)
```

```python
subjects = ["Machine Learning","AI","Python","CloudServices","Database","Prompt Engineering"]
most_toppers = []

for i in range(6):
    scores = [marks[i] for _, marks in passed]
    max_score = max(scores)
    count = sum(1 for _, marks in passed if marks[i] == max_score)
    most_toppers.append((subjects[i], count))

max_count = max(c for _, c in most_toppers)
subjects_with_most_toppers = [s for s, c in most_toppers if c == max_count]

print(subjects_with_most_toppers)
```

```python
fail_counts = []

for i in range(6):
    count = 0
    for _, _, marks in emp_records:
        if not (isinstance(marks[i], int) and marks[i] >= 35):
            count += 1
    fail_counts.append((subjects[i], count))

max_fail = max(c for _, c in fail_counts)
subjects_with_most_fails = [s for s, c in fail_counts if c == max_fail]

print(subjects_with_most_fails)
```
